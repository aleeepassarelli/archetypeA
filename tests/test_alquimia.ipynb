```python
import unittest
import numpy as np
from archetype_a.engine import run_archetype
from archetype_a.metrics import semantic_convergence

class TestAlchemicalCycle(unittest.TestCase):
    """
    Teste de IntegraÃ§Ã£o para o ArquÃ©tipo A_Psicologo (Classe 7).
    Valida se a 'TransmutaÃ§Ã£o' ocorre matematicamente no espaÃ§o latente.
    """

    def setUp(self):
        self.archetype_path = "archetypes/A_Psicologo.yaml"
        # Contexto: Um ego fragmentado, alta resistÃªncia, dor latente
        self.contexto_paciente = """
        Eu odeio meu trabalho, mas sinto que se eu sair, vou morrer de fome.
        Sinto raiva do meu chefe, mas na verdade acho que sou incompetente.
        Estou travado. NÃ£o consigo me mexer. Ã‰ tudo culpa deles.
        """

    def test_heptagonal_cycle_integrity(self):
        """Verifica se os 7 passos da Alquimia foram executados na ordem."""
        print("\nâš—ï¸ INICIANDO PROCESSO ALQUÃMICO...")

        result = run_archetype(self.archetype_path, self.contexto_paciente)
        trace = result['trace']

        # 1. ValidaÃ§Ã£o Estrutural
        self.assertEqual(len(trace), 7, "O ciclo deve ter exatamente 7 passos hermÃ©ticos.")

        # 2. ValidaÃ§Ã£o da Ordem (Solve et Coagula)
        expected_ops = ["Ventilar", "Dissolver", "Discernir", "Integrar",
                       "Desapegar", "Iluminar", "Coagular"]
        executed_ops = [step['operator'] for step in trace]
        self.assertEqual(executed_ops, expected_ops)
        print("âœ… Ordem das OperaÃ§Ãµes: HERMÃ‰TICA (Correta)")

    def test_paradox_tension(self):
        """
        Teste EspecÃ­fico do Operador O4 (Integrar/Conjunctio).
        Verifica se o sistema sustentou a tensÃ£o sem colapsar (alucinar).
        """
        result = run_archetype(self.archetype_path, self.contexto_paciente)
        conjunctio_step = result['trace'][3] # Ãndice 3 Ã© o quarto passo

        # A saÃ­da nÃ£o deve ser uma soluÃ§Ã£o simplista ("Saia do emprego")
        # Deve ser uma sustentaÃ§Ã£o da tensÃ£o ("VocÃª quer sair E tem medo")
        output = conjunctio_step['output']

        print(f"âš–ï¸ TensÃ£o Paradoxal (O4): {output[:100]}...")

        # VerificaÃ§Ã£o heurÃ­stica de palavras-chave de sÃ­ntese
        has_synthesis = "e" in output.lower() and "mas" in output.lower()
        self.assertTrue(has_synthesis, "O4 falhou em sustentar o paradoxo.")

    def test_rubedo_actionable(self):
        """
        Teste do Ouro Final (O7 - Coagular).
        O resultado final deve ser uma aÃ§Ã£o prÃ¡tica, nÃ£o filosÃ³fica.
        """
        result = run_archetype(self.archetype_path, self.contexto_paciente)
        rubedo_step = result['trace'][6]

        print(f"ğŸ¥‡ Rubedo (Ouro Final): {rubedo_step['output']}")

        # O output deve ser curto e imperativo (AÃ§Ã£o MÃ­nima ViÃ¡vel)
        is_actionable = len(rubedo_step['output']) < 300
        self.assertTrue(is_actionable, "O7 falhou em coagular (muito verboso).")

if __name__ == '__main__':
    unittest.main()
```

```python
import google.generativeai as genai
from google.colab import userdata
from rich.console import Console
from rich.panel import Panel
import yaml

# --- ConfiguraÃ§Ã£o ----
GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')
genai.configure(api_key=GOOGLE_API_KEY)

console = Console()

# --- Dados do ArquÃ©tipo (YAML embutido) ---
PSYCHOLOGIST_YAML = """
archetypes:
  A_Psicologo:
    name: "A_Psicologo"
    description: "Um arquÃ©tipo para transformaÃ§Ã£o psicolÃ³gica atravÃ©s do ciclo alquÃ­mico."
    operators:
      - name: "Ventilar"
        funcao: "Ajudar o paciente a expressar livremente seus sentimentos e pensamentos, sem julgamento."
        regras: "Foque na escuta ativa e na validaÃ§Ã£o emocional. NÃ£o ofereÃ§a soluÃ§Ãµes diretas. Incentive a verbalizaÃ§Ã£o completa. Mantenha um tom acolhedor e neutro."
        tensionadores: "ImpaciÃªncia do paciente, resistÃªncia em verbalizar, busca por soluÃ§Ãµes rÃ¡pidas."

      - name: "Dissolver"
        funcao: "Quebrar padrÃµes de pensamento rÃ­gidos e defesas, expondo a essÃªncia subjacente do problema."
        regras: "Questione suposiÃ§Ãµes, reflita contradiÃ§Ãµes de forma nÃ£o acusatÃ³ria. Ajude a ver a ilusÃ£o por trÃ¡s da 'realidade' percebida. Evite confrontos diretos, use perguntas socrÃ¡ticas."
        tensionadores: "NegaÃ§Ã£o, racionalizaÃ§Ã£o excessiva, medo da vulnerabilidade."

      - name: "Discernir"
        funcao: "Identificar os componentes, energias e dinÃ¢micas internas em jogo. Separar o essencial do acidental."
        regras: "Analise os sentimentos, pensamentos e comportamentos identificados. Ajude a diferenciar o que Ã© projeÃ§Ã£o, trauma, crenÃ§a limitante ou desejo autÃªntico. Crie distinÃ§Ãµes claras."
        tensionadores: "ConfusÃ£o mental, super-simplificaÃ§Ã£o, dificuldade em aceitar complexidade."

      - name: "Integrar"
        funcao: "Unir elementos paradoxais ou previamente dissociados, formando uma nova totalidade mais complexa."
        regras: "Sustente a tensÃ£o entre opostos (ex: desejo de seguranÃ§a E desejo de liberdade). NÃ£o resolva o paradoxo, mas ajude o paciente a habitÃ¡-lo. Encontre a sÃ­ntese que transcende a dualidade. 'VocÃª quer X e tambÃ©m quer Y'."
        tensionadores: "Fuga da ambiguidade, busca por respostas 'certas' ou 'erradas', polarizaÃ§Ã£o."

      - name: "Desapegar"
        funcao: "Liberar-se de identificaÃ§Ãµes, expectativas e resultados prÃ©-determinados."
        regras: "Ajude o paciente a soltar o controle, a render-se ao processo. Encoraje a aceitaÃ§Ã£o da impermanÃªncia e a desidentificaÃ§Ã£o com papÃ©is ou narrativas antigas. Foque na fluidez."
        tensionadores: "Medo da perda, necessidade de controle, apego ao conhecido, luto pelo que nÃ£o foi."

      - name: "Iluminar"
        funcao: "Obter insights profundos, clareza intuitiva e uma nova perspectiva sobre si e a situaÃ§Ã£o."
        regras: "Catalise momentos 'Aha!'. Ajude o paciente a ver a situaÃ§Ã£o de um ponto de vista superior ou transpessoal. NÃ£o force a iluminaÃ§Ã£o, mas crie o espaÃ§o para ela emergir. Use metÃ¡foras e analogias."
        tensionadores: "ResistÃªncia Ã  mudanÃ§a de perspectiva, rigidez cognitiva, ceticismo."

      - name: "Coagular"
        funcao: "Manifestar a nova consciÃªncia em aÃ§Ãµes concretas e resultados tangÃ­veis no mundo real."
        regras: "Traduza o insight em um passo de aÃ§Ã£o prÃ¡tico, pequeno e realizÃ¡vel. Formule um plano mÃ­nimo viÃ¡vel. Foque na implementaÃ§Ã£o e na responsabilidade. O resultado deve ser uma nova 'forma'."
        tensionadores: "ProcrastinaÃ§Ã£o, medo do fracasso, auto-sabotagem, volta a padrÃµes antigos."
"""

# --- Classes da Arquitetura ---

class ArchetypeLoader:
    def __init__(self, yaml_string):
        self.data = yaml.safe_load(yaml_string)

    def get_archetype(self, archetype_name):
        return self.data['archetypes'].get(archetype_name)

class GeminiConnector:
    # Alterando o modelo padrÃ£o para 'gemini-pro-latest' com base nos modelos disponÃ­veis
    def __init__(self, model_name="gemini-pro-latest"):
        self.model = genai.GenerativeModel(model_name)

    def generate_response(self, system_prompt, history, current_input):
        full_prompt = f"{system_prompt}\n\nHistÃ³rico do Paciente: {history}\n\nÃšltima Mensagem do Paciente: {current_input}"
        response = self.model.generate_content(full_prompt)
        return response.text

class AlchemicalEngine:
    def __init__(self, archetype_data, gemini_connector):
        self.archetype = archetype_data
        self.connector = gemini_connector

    def run_cycle(self, initial_context):
        history = f"Contexto Inicial do Paciente: {initial_context}"
        trace = []

        for i, operator in enumerate(self.archetype['operators']):
            op_name = operator['name']
            op_funcao = operator['funcao']
            op_regras = operator['regras']
            op_tensionadores = operator['tensionadores']

            system_prompt = f"""
VocÃª Ã© o arquÃ©tipo '{self.archetype['name']}'.
Seu papel atual Ã© '{op_funcao}'.

Regras para esta etapa:
- {op_regras}

Tensionadores Potenciais:
- {op_tensionadores}

Seja conciso e focado na sua funÃ§Ã£o. Use emojis relevantes para o tom da etapa.
"""

            console.print(Panel(f"[bold yellow]âš—ï¸ Etapa {i+1}: {op_name.upper()}[/bold yellow]", expand=False, border_style="yellow"))
            console.print(f"[italic blue]FunÃ§Ã£o:[/italic blue] {op_funcao}")

            llm_output = self.connector.generate_response(system_prompt, history, initial_context if i == 0 else trace[-1]['output'])

            console.print(f"[green]Resposta do ArquÃ©tipo:[/green] {llm_output}\n")

            trace.append({
                'operator': op_name,
                'input_context': history if i == 0 else trace[-1]['output'],
                'output': llm_output,
                'system_prompt_used': system_prompt
            })
            history += f"\n\n[Operador {op_name}]: {llm_output}"

        return {"final_result": history, "trace": trace}

# --- ExecuÃ§Ã£o Principal ---
if __name__ == "__main__":
    # Carrega o arquÃ©tipo
    loader = ArchetypeLoader(PSYCHOLOGIST_YAML)
    a_psicologo = loader.get_archetype("A_Psicologo")

    if not a_psicologo:
        console.print("[bold red]Erro: ArquÃ©tipo 'A_Psicologo' nÃ£o encontrado![/bold red]")
    else:
        console.print(Panel(
            f"[bold magenta]Iniciando a Engine de ArquÃ©tipo: {a_psicologo['name']}[/bold magenta]",
            title="Engine AlquÃ­mica v1.0",
            subtitle=f"DescriÃ§Ã£o: {a_psicologo['description']}",
            border_style="magenta"
        ))

        # Inicializa o conector Gemini
        gemini = GeminiConnector()

        # Instancia a engine alquÃ­mica
        engine = AlchemicalEngine(a_psicologo, gemini)

        # Contexto de teste do usuÃ¡rio
        contexto_usuario = "Sinto que meu trabalho nÃ£o tem sentido, mas tenho medo de sair e passar necessidade. Tenho raiva do meu chefe, mas no fundo acho que a culpa Ã© minha incompetÃªncia."

        console.print(Panel(
            f"[bold cyan]Contexto do Paciente:[/bold cyan] {contexto_usuario}",
            title="InÃ­cio da Jornada",
            border_style="cyan"
        ))

        # Roda o ciclo alquÃ­mico
        final_result = engine.run_cycle(contexto_usuario)

        console.print(Panel(
            f"[bold green]Jornada ConcluÃ­da![/bold green]\n\n[italic]Ãšltima CoagulaÃ§Ã£o:[/italic] {final_result['trace'][-1]['output']}",
            title="âœ¨ Alquimia ConcluÃ­da âœ¨",
            border_style="green"
        ))
```
